## Steps for applying an algorithm

#### Spliting the dataset in independent and dependent variables

```python
x=iris.iloc[:,:4].values
y=iris["Species"].values
```

#### Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

#### Feature Scaling

```python
print("*"*120)
print("STANDARDIZING THE DATASET".center(50))
print("*"*120)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

#### TRAIN THE MODEL USING SVC WITH LINEAR KERNEL

**Here we are using SVC classifier to predict the accuracy of the model**

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="linear", random_state=0)
```

```python
# Train the algorithm by using training input and training output
svc.fit(x_train, y_train)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

#### Predicting the Test set results

```python
y_pred = svc.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

#### Model Evaluation

```python
print("*"*120)
print("ACCURACY SCORE OF THE MODEL USING SVC ALGORITHM".center(100))
print("*"*120)
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
acc_svc = accuracy_score(y_test, y_pred)
print(acc_svc)
```

```python
print("*"*120)
print("PLOT FOR THE CONFUSION MATRIX OF THE MODEL USING SVC ALGORITHM".center(100))
print("*"*120)

from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(svc, x_test, y_test)
```

```python
print("Confusion Matrix:\n")
Matrix(confusion_matrix(y_test,y_pred))
```

```python
print("*"*120)
print("CLASSIFICATION REPORT OF THE MODEL USING SVC ALGORITHM".center(100))
print("*"*120)

class_report = classification_report(y_test, y_pred)

print(class_report)
```
